```python
# function to create a dictionary of unique values of column mapped to numerical values
def getMappingDictionary(unique_data):
    mappingDictionary = {}
    for i in range(len(unique_data)):
        mappingDictionary[unique_data[i]] = i
    return mappingDictionary
```

```python
def getUniqueDataList(columnName):
    return list(set(dataset[columnName]))
```

```python
def modifyDatasetColumn(columnName, mapping_dict):
    dataset[columnName] = dataset[columnName].map(mapping_dict)
```

```python
def modifyDataset(columnName):
    unique_columnValues = getUniqueDataList(columnName)
    print(unique_columnValues)
    mapping_dict = getMappingDictionary(unique_columnValues)
    print(mapping_dict)
    modifyDatasetColumn(columnName,mapping_dict)
```

```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
import pandas as pd

# Assume df is your dataframe
# Apply one-hot encoding to all categorical columns
df_encoded = pd.get_dummies(df, drop_first=True)

# Create a KMeans model and use the KElbowVisualizer
model = KMeans()
visualizer = KElbowVisualizer(model, k=(1, 10))

# Fit the data to the visualizer
visualizer.fit(df_encoded)

# Display the elbow plot
visualizer.show()

# Get the optimal number of clusters (elbow value)
k = visualizer.elbow_value_
print("Optimal number of clusters:", k)
```

```python
# Initialize KMeans model
kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)

# Fit the KMeans model and predict cluster labels
customer_segments = kmeans.fit_predict(df_encoded)

# Add cluster labels to the dataframe
df_encoded['cluster'] = customer_segments
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
import joblib  # For saving and loading the column names (optional)

# Sample input data for the new customer
input_customer_data = {
    'age': [40],
    'gender': ['Female'],
    'relationship': ['Colleague'],
    'occasion': ['Karva Chauth'],
    'budget': [10000],
    'interests': ['Photography,Astronomy']
}

# Convert to DataFrame
new_customer = pd.DataFrame(input_customer_data)

# Split 'interests' and use MultiLabelBinarizer
new_customer['interests'] = new_customer['interests'].apply(lambda x: x.split(','))
mlb = MultiLabelBinarizer()
interests_encoded = pd.DataFrame(mlb.fit_transform(new_customer['interests']), columns=mlb.classes_)

# Drop original interests and add encoded ones
new_customer = pd.concat([new_customer.drop('interests', axis=1), interests_encoded], axis=1)

# One-hot encode categorical columns (gender, relationship, occasion)
new_customer_encoded = pd.get_dummies(new_customer)

# Option 1: Load saved column names from training data (recommended approach)
try:
    expected_columns = joblib.load('kmeans_feature_columns.pkl')  # Load saved columns
except FileNotFoundError:
    # Hardcoded list of expected columns (for testing purposes)
    expected_columns = [
        'age', 'budget', 'gender_Female', 'relationship_Colleague', 
        'occasion_Karva Chauth', 'Photography', 'Astronomy'
    ]

# Align new customer data with the training data columns
# Add missing columns with value 0
for col in expected_columns:
    if col not in new_customer_encoded.columns:
        new_customer_encoded[col] = 0

# Drop extra columns (if any)
new_customer_encoded = new_customer_encoded[expected_columns]

# Ensure new_customer_encoded is a 2D array (needed for prediction)
new_customer_encoded = new_customer_encoded.values.reshape(1, -1)

# Assuming `kmeans` is your trained model
# Predict the cluster for the new customer
new_customer_segment = kmeans.predict(new_customer_encoded)

# Output the result
print('New customer belongs to segment:', new_customer_segment[0])
```
